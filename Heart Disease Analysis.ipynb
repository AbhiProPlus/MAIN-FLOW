Here’s a detailed explanation of each column in the heart disease dataset:

1. age
Description: Age of the patient in years.
Type: Numerical (integer).
Range: 29–77 years.
Relevance: Age is a key risk factor for heart disease. Older individuals are generally at higher risk.

2. sex
Description: Gender of the patient.
0: Female
1: Male
Type: Categorical (binary).
Relevance: Gender influences the prevalence and risk of heart disease. Males typically have a higher risk.

3. cp (Chest Pain Type)
Description: Type of chest pain experienced by the patient.
0: Typical angina (related to reduced blood flow to the heart).
1: Atypical angina.
2: Non-anginal pain.
3: Asymptomatic (no chest pain).
Type: Categorical.
Relevance: Chest pain type is a strong indicator of heart-related issues.

4. trestbps (Resting Blood Pressure)
Description: Resting blood pressure in mm Hg (measured at admission).
Type: Numerical (integer).
Range: 94–200 mm Hg.
Relevance: High blood pressure (hypertension) is a major risk factor for heart disease.

5. chol (Serum Cholesterol)
Description: Serum cholesterol level in mg/dL.
Type: Numerical (integer).
Range: 126–564 mg/dL.
Relevance: High cholesterol can lead to plaque buildup in arteries, increasing heart disease risk.

6. fbs (Fasting Blood Sugar)
Description: Indicates if fasting blood sugar is greater than 120 mg/dL.
0: Fasting blood sugar ≤ 120 mg/dL.
1: Fasting blood sugar > 120 mg/dL.
Type: Categorical (binary).
Relevance: Elevated fasting blood sugar is associated with diabetes, which increases the risk of heart disease.

7. restecg (Resting Electrocardiographic Results)
Description: Results of the resting ECG.
0: Normal.
1: ST-T wave abnormality (indicative of potential heart issues).
2: Left ventricular hypertrophy (LVH).
Type: Categorical.
Relevance: Abnormal ECG results can indicate heart abnormalities.

8. thalach (Maximum Heart Rate Achieved)
Description: Maximum heart rate achieved during a stress test.
Type: Numerical (integer).
Range: 71–202 bpm.
Relevance: A lower maximum heart rate can indicate heart issues.

9. exang (Exercise-Induced Angina)
Description: Indicates whether the patient experiences angina during exercise.
0: No.
1: Yes.
Type: Categorical (binary).
Relevance: Exercise-induced angina suggests reduced blood flow to the heart during exertion.

10. oldpeak
Description: ST depression induced by exercise relative to rest.
Type: Numerical (float).
Range: 0.0–6.2.
Relevance: Measures changes in the heart's electrical activity; abnormal values can indicate ischemia.

11. slope (Slope of the ST Segment)
Description: The slope of the ST segment during peak exercise.
0: Upsloping.
1: Flat.
2: Downsloping.
Type: Categorical.
Relevance: ST segment slope can indicate heart health during stress testing.

12. ca (Number of Major Vessels Colored by Fluoroscopy)
Description: Number of major blood vessels visible under fluoroscopy (0–4).
Type: Numerical (integer).
Relevance: Indicates blockages in major vessels; higher numbers suggest worse conditions.

13. thal (Thalassemia Status)
Description: Blood disorder indicator.
0: Unknown.
1: Normal.
2: Fixed defect (permanent damage to the heart muscle).
3: Reversible defect (temporary damage or blockage).
Type: Categorical.
Relevance: Indicates heart health issues related to blood supply.

14. target
Description: Presence or absence of heart disease (dependent variable).
0: No heart disease.
1: Heart disease present.
Type: Categorical (binary).
Relevance: This is the variable to predict using other features.

 IQR method to detect and handle outliers

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = ((df < lower_bound) | (df > upper_bound))

# Count number of outliers in each column
outliers_count = outliers.sum()
print(outliers_count)

```

```python
# Optionally, remove outliers
df_cleaned = df[~((df < lower_bound) | (df > upper_bound)).any(axis=1)]
```

Look at the patterns of blue and orange dots:
Do they form distinct clusters? If yes, these feature pairs might be great for separating target=0 and target=1.
If the colors mix randomly, it means the combination of those features doesn't differentiate well between the two targets.

6) Max Heart Rate: Is there a significant difference in maximum heart rate achieved (thalach) between individuals with and without heart disease?

```python
# Descriptive statistics for max heart rate (thalach) by target (heart disease status)
thalach_stats = df.groupby('target')['thalach'].describe()
print(thalach_stats)
```

The descriptive statistics for thalach (maximum heart rate) indicate the following:

Without heart disease (target = 0):

Mean: 139.13
Standard Deviation: 22.57
Min: 71
Max: 195
With heart disease (target = 1):

Mean: 158.59
Standard Deviation: 19.10
Min: 96
Max: 202

Analysis:
Higher mean: The individuals with heart disease tend to have a higher average maximum heart rate (thalach).
Range: The range of heart rates is similar, but the group with heart disease has a wider spread (higher max value).

```python
from scipy.stats import ttest_ind

# Separate the data for individuals with and without heart disease
with_heart_disease = df[df['target'] == 1]['thalach']
without_heart_disease = df[df['target'] == 0]['thalach']

# Perform the independent t-test
t_stat, p_value = ttest_ind(with_heart_disease, without_heart_disease)

# Display the result
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Interpretation
if p_value < 0.05:
    print("The difference in maximum heart rate between individuals with and without heart disease is statistically significant.")
else:
    print("The difference in maximum heart rate is not statistically significant.")
```

The result of the t-test indicates that the difference in maximum heart rate (thalach) between individuals with and without heart disease is statistically significant. Since the p-value is extremely small (much less than 0.05), we can confidently conclude that people with heart disease tend to have a different (likely lower) maximum heart rate compared to those without heart disease.

To summarize:
T-statistic: 14.93
P-value: 9.96e-46 (very small)
Conclusion: The difference in maximum heart rate between the two groups is statistically significant.
